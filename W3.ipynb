### About functions

***

```python
def square(n):
    ans = n * n
    return ans
```

```python
square (4)
```

```python
def power(n, m):
    ans = 1 
    for i in range(m):
        ans = ans * n
    return ans
```

```python
power(4, 2)
```

### Map

```python
numbers = list(range(10))
```

```python
numbers
```

```python
def mapall(f, l):
    return[f(i) for i in l]
```

```python
mapall(square, numbers)
```

```python
mapall(str, numbers)
```

```python
mapall(lambda x:x-5, numbers)
```

## Models

```python
def f(x, p):
    return p[0] + x * p[1]
```

### Anaysis

```python
import seaborn as sns
#load dataset.
penguins = sns.load_dataset("penguins")
# have a look at it.
sns.pairplot(penguins, hue="species")
```

```python
#pick out two variables.
flipper = penguins[["body_mass_g", "flipper_length_mm"]].dropna()
# Scatter and fit line for just those two  variables.
sns.regplot(x="body_mass_g", y="flipper_length_mm", data=penguins)
```

```python
import sklearn.linear_model as lin
x = flipper["body_mass_g"].to_numpy()
y = flipper["flipper_length_mm"].to_numpy()

x = x.reshape(-1,1)

model = lin.LinearRegression()
model.fit(x, y)
r = model.score(x, y)
p =  [model.intercept_, model.coef_[0]]
```

```python
r
```

```python
p
```

### Predict

```python
f(4500.0, p)
```

```python
def predict(x):
    return f(x, p)
```

```python
predict(4500.0)
```
